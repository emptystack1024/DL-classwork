```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import copy
from lxml.etree import Element, SubElement, tostring, ElementTree

import xml.etree.ElementTree as ET
import pickle
import os
from os import listdir, getcwd
from os.path import join

classes = {'Truck': 0, 'car': 1, 'van': 2, 'Bus': 3, 'SUV': 0, 'Sedan': 1, 'Microbus': 3, 'Minivan': 2} # SUV 与 truck 为同一类别


def convert(size, box):
    dw = 1. / size[0]
    dh = 1. / size[1]
    x = (box[0] + box[1]) / 2.0    # (x_min + x_max) / 2.0
    y = (box[2] + box[3]) / 2.0    # (y_min + y_max) / 2.0
    w = box[1] - box[0]   # x_max - x_min
    h = box[3] - box[2]   # y_max - y_min
    x = x * dw
    w = w * dw
    y = y * dh
    h = h * dh
    return (x, y, w, h)


def convert_annotation(image_id):
    in_file  = open('/home/ubuntu/下载/dataset/BIT-Vehicle/annotations/%s.xml' % (image_id), encoding='UTF-8')

    out_file = open('/home/ubuntu/下载/dataset/BIT-Vehicle/labels/%s.txt' % (image_id), 'w')  # 生成txt格式文件
    tree = ET.parse(in_file)
    root = tree.getroot()
    size = root.find('size')
    w = size.find('width').text
    # 删除w中的[和]
    w = w.replace('[', '')
    w = w.replace(']', '')
    w = int(w)
    h = size.find('height').text
    h = h.replace('[', '')
    h = h.replace(']', '')
    h = int(h)

    for obj in root.iter('object'):
        cls = obj.find('name').text
        # print(cls)
        if cls not in classes:
            print(cls)
        cls_id = classes[cls]
        xmlbox = obj.find('bndbox')
        b = (float(xmlbox.find('xmin').text), float(xmlbox.find('xmax').text), float(xmlbox.find('ymin').text),
             float(xmlbox.find('ymax').text))
        bb = convert((w, h), b)
        out_file.write(str(cls_id) + " " + " ".join([str(a) for a in bb]) + '\n')

xml_path = '/home/ubuntu/下载/dataset/BIT-Vehicle/annotations'

# xml list
img_xmls = os.listdir(xml_path)
for img_xml in img_xmls:
    label_name = img_xml.split('.')[0]
    print(label_name)
    convert_annotation(label_name)
```
