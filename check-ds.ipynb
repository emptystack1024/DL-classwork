### 问题2
学会了如何读取图片和label信息，将其会知道一张图片中并且打印出来

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def draw_boxes(img_path, label_path):
    # 读取图像
    img = cv2.imread(img_path)
    # 读取标注文件
    with open(label_path, 'r') as file:
        lines = file.readlines()
    
    for line in lines:
        class_id, x_center, y_center, width, height = map(float, line.strip().split())
        # 计算边界框的左上角和右下角坐标
        x1 = int((x_center - (width / 2)) * img.shape[1])
        y1 = int((y_center - (height / 2)) * img.shape[0])
        x2 = int((x_center + (width / 2)) * img.shape[1])
        y2 = int((y_center + (height / 2)) * img.shape[0])

        # 在边框上打印类别
        cv2.putText(img, str(int(class_id)), (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
        
        # 绘制边界框
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
    
    # 在命令行中输出图像
    # 设置画布大小
    plt.figure(figsize=(20, 20))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()


# Stanford Cars
# image_path = '/home/ubuntu/下载/dataset/archive/img_test/000048.jpg'
# label_path = '/home/ubuntu/下载/dataset/archive/test_label/000048.txt'

# BITVehicle_Dataset
# image_path = '/home/ubuntu/下载/dataset/BIT-Vehicle/BITVehicle_Dataset/vehicle_0000002.jpg'
# label_path = '/home/ubuntu/下载/dataset/BIT-Vehicle/labels/vehicle_0000002.txt'

# UA-DETRAC
image_path = '/home/ubuntu/文档/GitHub/DL大作业/datasets/content/images/val/MVI_39031_img00001.jpg'
label_path = '/home/ubuntu/文档/GitHub/DL大作业/datasets/content/labels/val/MVI_39031_img00001.txt'


draw_boxes(image_path, label_path)
```
